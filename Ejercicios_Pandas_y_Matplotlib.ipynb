```python
x = [2,2,4,6,7,7,8,10,12,14]
y = [110,105,99,97,94,88,89,85,86,88]

fig, ax = plt.subplots()

ax.scatter(x, y, label='Autos', color='red', marker='o')
ax.set_title("Velocidad y edad de Autos", color='red', size=10)
ax.set_ylabel("Velocidad", color='red', size=10)
ax.set_xlabel("Edad", color='red', size=10)

"------------"

ax.set_xlim(0, 18)
#Se puede hacer ax.set_xlim(18,0) y te da el gráfico invertido
ax.grid(color = 'gray', linestyle = 'dashed')
#ax.set_xticks([0,5,10,15,20])
#ax.set_xlabel('Cualquier cosa')

ax.legend()
```

# **Ejercicio 8 - Grillas - Ej 2**

```python
import numpy as np
```

```python
np.random.seed(0)
df = pd.DataFrame(data={'a':np.random.randint(0, 100, 50),
                        'b':np.random.randint(0, 100, 50),
                        'c':np.random.randint(0, 100, 50),
                        'd':np.random.randint(0, 100, 50)})
df.head()
```

```python
x = df.index.values

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(15, 8))
fig.subplots_adjust(wspace=0.5,hspace=0.5)

ax[0].set_title('Grafico 1')
ax[0].set_xlabel('Eje X')
ax[0].set_ylabel('Eje Y')
ax[0].grid()
ax[0].plot(x, df['a'], label='A', color='green', linestyle='solid')
ax[0].plot(x, df['b'], label='B', color='red', linestyle='dashed')
ax[0].legend()

ax[1].set_title('Grafico 2')
ax[1].set_xlabel('Eje X')
ax[1].set_ylabel('Eje Y')
ax[1].grid()
ax[1].scatter(x, df['c'], label='C', color='magenta', marker='o')
ax[1].scatter(x, df['d'], label='D', color='blue', marker='v')
ax[1].legend()

plt.show()
```

```python
x = df.index.values

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15, 8))
fig.subplots_adjust(wspace=0.5,hspace=0.5)

ax[0, 0].set_title('Grafico 1')
ax[0, 0].set_xlabel('Eje X')
ax[0, 0].set_ylabel('Eje Y')
ax[0, 0].grid()
ax[0, 0].plot(x, df['a'], label='A', color='green', linestyle='solid')
ax[0, 0].plot(x, df['b'], label='B', color='red', linestyle='dashed')
ax[0, 0].legend()

ax[1, 1].set_title('Grafico 2')
ax[1, 1].set_xlabel('Eje X')
ax[1, 1].set_ylabel('Eje Y')
ax[1, 1].grid()
ax[1, 1].scatter(x, df['c'], label='C', color='magenta', marker='o')
ax[1, 1].scatter(x, df['d'], label='D', color='blue', marker='v')
ax[1, 1].legend()

plt.show()
```
