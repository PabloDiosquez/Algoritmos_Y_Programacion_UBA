## Ejercicio de tipo de gráficos

Para realizar este ejercicio, debemos importar la información del PBI per cápita de los distintos países, a lo largo de un período que abarca desde 1952 y 2007.

```python
# Importo la información
url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminder_with_codes.csv"
data = pd.read_csv(url)

# Modificar el tipo de dato:
data['year'] = data['year'].astype("int")

data.head()
```

```python
data.info()
```

**Ejercicio 1:**

Elije un año en el que desees ver la relación entre la expectativa de vida de los habitantes (columna `lifeExp`) y el PBI per cápita de los habitantes (columna gdpPercap)

```python
# Modifica este valor
# =========== Código de alumno ===============
year = 1972
# ============================================

data_year = data[data["year"] == year]
data_year.head()
```

Realiza un gráfico de puntos que muestre la relación entre la expectativa de vida (columna `lifeExp`) y el PBI per cápita de los habitantes (columna gdpPercap).

El gráfico debe tener:
- Título apropiado
- Nombre y unidades de los ejes cartesianos
- Marcador de tipo triangular y color "#23A763"
- Grilla

```python
# Crear una figura y un eje (subplot) para el gráfico de dispersión
fig, ax = plt.subplots()

# Extraer los datos de expectativa de vida y PBI per cápita
expectativa_de_vida = data_year['lifeExp']
pbi_per_capita = data_year['gdpPercap']

# Configurar el título y las etiquetas de los ejes
ax.set_title("Relación entre Expectativa de Vida y PBI per Cápita (Año 1972)")
ax.set_xlabel("Expectativa de Vida (en años)")
ax.set_ylabel("PBI per Cápita")

# Establecer una línea de grilla con estilo discontinuo
ax.grid(color='gray', linestyle='dashed')

# Crear un gráfico de dispersión para mostrar la relación
ax.scatter(expectativa_de_vida, pbi_per_capita, marker='^', color='#23A763')

# Mostrar el gráfico
plt.show()
```

**Hagamos algo mejor 🎉**

```python
def create_easy_graph(x, y, label, ax, xlabel, ylabel, title, color):
  """
  Crea un gráfico a partir de vectores con valores de los ejes x e y.
  Parámetros:
    - El texto para el label.
    - El subplot donde graficar.
    - Un label para el eje x.
    - Un label para el eje y.
    - Un título para el gráfico.
    - Un color.
  Observaciones:
    - El gráfico y el color pueden ser nulos. En ese caso, toman valores por defecto.
  """
  if ax is None:
    fig, ax = plt.subplots()

  if color is None:
    color = 'black'

  # Se define el gráfico
  ax.plot(x, y, label=label, color=color)
  ax.set_title(title)
  ax.set_xlabel(xlabel)
  ax.set_ylabel(ylabel)

  return ax
```

```python
# Crea una figura con una cuadrícula de 2x2 de subgráficos y establece el tamaño
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15, 8))

# Nombres de las columnas que se utilizarán en los gráficos y sus colores correspondientes
columnas = ['a', 'b', 'c', 'd']
colores = ['green', 'red', 'magenta', 'blue']

# Itera sobre las filas y columnas de la cuadrícula de subgráficos
for fila in range(2):
    for columna in range(2):
        ax[fila, columna].grid(color='gray', linestyle='--')

        # Reemplaza 'create_easy_graph' con la función que utilizas para crear gráficos
        create_easy_graph(df.index.values, df[columnas[2*fila + columna]], f'Columna {columnas[2*fila + columna]}', ax[fila, columna], 'Índices', 'Valores', f'Columna "{columnas[2*fila + columna]}"', colores[2*fila + columna])

# Ajusta el espaciado entre los subgráficos
plt.tight_layout()

# Muestra los gráficos
plt.show()
```
