```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph ###

```python
# Defining the data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Resize the graph
plt.figure(figsize=(6, 5))

# First plot: f(x) = x^2 with different line styles and markers
plt.plot(x[:4], y[:4], 'ro--', label='f(x) = x^2')  # Red line with circles, dashed line
plt.plot(x[3:], y[3:], 'bo-', label='f(x) = x^2')   # Blue line with circles, solid line

# Second plot: g(x) = 3x with green dashed lines
xs = np.arange(0, 5.5, 0.5)
plt.plot(xs, 3*xs, 'go--', label='g(x) = 3x')

# Title and axis labels configuration
plt.title('Our First Graph!')  # Graph title
plt.xlabel('X Axis')           # X-axis label
plt.ylabel('Y Axis')           # Y-axis label

# Configuration of ticks and labels on the axes
plt.xticks([x for x in range(6)])           # Ticks on the X-axis (0 to 5)
plt.yticks([x for x in range(0, 31, 5)])    # Ticks on the Y-axis (0 to 30, in increments of 5)

# Show the legend indicating which line corresponds to which function
plt.legend()

# Display the graph
plt.show()
```

### Bar Chart ###

```python
# Define the labels and values for the bar chart
labels = ['A', 'B', 'C']
values = [10, 20, 30]

# Create a figure with a specified size
plt.figure(figsize=(7, 6))

# Create a bar chart with labels and values
bars = plt.bar(labels, values)

# Uncomment and use the lines below to add patterns to the bars
# bars[0].set_hatch('/')  # Add hatch pattern to the first bar
# bars[1].set_hatch('o')  # Add hatch pattern to the second bar
# bars[2].set_hatch('*')  # Add hatch pattern to the third bar

# Define a list of hatch patterns
patterns = ['/', 'o', '*']

# Loop through the bars and assign hatch patterns
for index, bar in enumerate(bars):
    bar.set_hatch(patterns[index])

# Display the bar chart
plt.show()
```
