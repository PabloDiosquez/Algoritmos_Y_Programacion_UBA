**Load Necessary Libraries**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph ###

```python
# Defining the data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Resize the graph
plt.figure(figsize=(6, 5))

# First plot: f(x) = x^2 with different line styles and markers
plt.plot(x[:4], y[:4], 'ro--', label='f(x) = x^2')  # Red line with circles, dashed line
plt.plot(x[3:], y[3:], 'bo-', label='f(x) = x^2')   # Blue line with circles, solid line

# Second plot: g(x) = 3x with green dashed lines
xs = np.arange(0, 5.5, 0.5)
plt.plot(xs, 3*xs, 'go--', label='g(x) = 3x')

# Title and axis labels configuration
plt.title('Our First Graph!')  # Graph title
plt.xlabel('X Axis')           # X-axis label
plt.ylabel('Y Axis')           # Y-axis label

# Configuration of ticks and labels on the axes
plt.xticks([x for x in range(6)])           # Ticks on the X-axis (0 to 5)
plt.yticks([x for x in range(0, 31, 5)])    # Ticks on the Y-axis (0 to 30, in increments of 5)

# Show the legend indicating which line corresponds to which function
plt.legend()

# Display the graph
plt.show()
```

### Bar Chart ###

```python
# Define the labels and values for the bar chart
labels = ['A', 'B', 'C']
values = [10, 20, 30]

# Create a figure with a specified size
plt.figure(figsize=(7, 6))

# Create a bar chart with labels and values
bars = plt.bar(labels, values)

# Uncomment and use the lines below to add patterns to the bars
# bars[0].set_hatch('/')  # Add hatch pattern to the first bar
# bars[1].set_hatch('o')  # Add hatch pattern to the second bar
# bars[2].set_hatch('*')  # Add hatch pattern to the third bar

# Define a list of hatch patterns
patterns = ['/', 'o', '*']

# Loop through the bars and assign hatch patterns
for index, bar in enumerate(bars):
    bar.set_hatch(patterns[index])

# Display the bar chart
plt.show()
```

### Real World Examples ###

```python
gas = pd.read_csv('gas_prices.csv')

data = pd.DataFrame(gas)

data.head()
```

### Line Graph ###

```python
# Create a figure with a specified size
plt.figure(figsize=(8, 5))

# Extract data for the USA and Canada
gas_usa = data['USA']
gas_canada = data['Canada']

# Plot the data for the USA and Canada with labels, colors, and markers
plt.plot(data['Year'], gas_usa, label='USA', color='r', marker='o')
plt.plot(data['Year'], gas_canada, label='Canada', color='g', marker='o')

# Set the title and y-axis label for the plot
plt.title('USA vs Canada Gas Prices')
plt.ylabel('Dollars/Gallon')

# Display a legend indicating which line corresponds to which country
plt.legend()

# Customize the x-axis ticks to show every third year
plt.xticks(data['Year'][::3])

# Add gridlines to the plot with specified color, linewidth, and linestyle
plt.grid(color='gray', linewidth=1, linestyle='--')

# Display the plot
plt.show()
```

### Histograms ###

```python
# Read the data from a CSV file named 'fifa_data.csv' using pandas
fifa_file = pd.read_csv('fifa_data.csv')

# Create a pandas DataFrame from the read data
fifa_df = pd.DataFrame(fifa_file)

# Display the contents of the DataFrame, showing the data from the CSV file
fifa_df.head(10)
```

```python
# Define the bins for the histogram. This will create bins ranging from 40 to 100 in steps of 10.
bins = [x for x in range(40, 101, 10)]

# Create a histogram of the 'Overall' column from the 'fifa_df' DataFrame using the specified bins,
# and set the color to a shade of blue (#5886e8).
plt.hist(fifa_df['Overall'], bins=bins, color='#5886e8')

# Set the title of the histogram.
plt.title('Distribution of Player Skills in FIFA 2018')

# Set the label for the y-axis, indicating the number of players.
plt.ylabel('Number of players')

# Set the label for the x-axis, indicating the skill level.
plt.xlabel('Skill level')

# Display the histogram plot.
plt.show()
```

### Pie Chart

```python
# Count the number of players who prefer the left foot
left = len(fifa_df.loc[fifa_df['Preferred Foot'] == 'Left'])

# Count the number of players who prefer the right foot
right = len(fifa_df.loc[fifa_df['Preferred Foot'] == 'Right'])

# Create a figure for the pie chart with custom dimensions and background color
plt.figure(figsize=(7, 6), facecolor='#dfe8ca')

# Labels for the two segments of the pie chart
labels = ['Left', 'Right']

# Colors for the two segments of the pie chart
colors = ['#5184f0', '#cf2d7b']

# Set the title of the chart
plt.title('Foot preference of FIFA Players')

# Create the pie chart with values of 'left' and 'right', using the defined labels and colors.
# Also, display the percentage inside the segments and adjust the visual appearance with wedgeprops.
plt.pie([left, right], labels=labels, colors=colors, autopct='%2.f%%', wedgeprops={"linewidth": 1, "edgecolor": "white"})

# Show the pie chart
plt.show()
```

```python
# Convert the 'Weight' column to integers by stripping 'lbs' and handling non-string values
fifa_df.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa_df.Weight]

# Create a figure for the pie chart with custom dimensions and background color
plt.figure(figsize=(7, 6), facecolor='#e3f0c5')

# Apply a specific style to the plot (ggplot style)
plt.style.use('ggplot')

# Count the number of players falling into different weight categories
light = len(fifa_df.loc[fifa_df.Weight < 125])
light_medium = len(fifa_df.loc[fifa_df.Weight.between(125, 150)])
medium = len(fifa_df.loc[fifa_df.Weight.between(150, 175)])
medium_heavy = len(fifa_df.loc[fifa_df.Weight.between(175, 200)])
heavy = len(fifa_df.loc[fifa_df.Weight >= 200])

# Create a list of counts for each weight category
weights = [light, light_medium, medium, medium_heavy, heavy]

# Labels for the weight categories
labels = ['Under 125', '125 - 150', '150 - 175', '175 - 200', 'Over 200']

# Specify the degree of separation (explode) for specific slices of the pie chart
explode = [0.4, 0.1, 0, 0, 0.4]

# Create the pie chart with the defined weights, labels, percentage format, separation, and visual properties
plt.pie(weights, labels=labels, autopct='%.2f%%', pctdistance=0.8, explode=explode, wedgeprops={"linewidth": 0.5, "edgecolor": "black"})

# Display the pie chart
plt.show()
```
