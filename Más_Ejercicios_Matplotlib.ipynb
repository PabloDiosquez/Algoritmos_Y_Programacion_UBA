```python
from matplotlib import pyplot as plt
```

```python
# Set de datos:
peliculas = {'nombre': ['Titanic', 'Kil Bill', 'Matrix', 'El padrino', 'Avatar',
                        'Casablanca', 'El exorcista',  'Soy leyenda',
                        'El club de la pelea', 'Mujercitas'],
            'director': ['James Cameron', 'Quentin Tarantino', 'Hermanas Wachowski',
                        'Francis Ford Coppola', 'James Cameron', 'Michael Curtiz',
                        'William Friedkin', 'Francis Lawrence','David Fincher',
                        'Greta Gerwig'],
            'año': [1997, 2003, 1999, 1972, 2009, 1942, 1973, 2007, 1999, 2019],
            'género': ['romance', 'acción', 'ciencia ficción', 'drama', 'ciencia ficción', 'drama', 'terror',
                        'ciencia ficción', 'drama', 'drama'],
            'puntaje': [8.6, None, 6.9, 7.5, 9.1, 6.0, None, None, 9.4, 8.0],
             'vista': ['si', 'no', 'si', 'si', 'si', 'si', 'no', 'no', 'si', 'si']
             }
```

**Ejercicio 1**

Graficar la función `f(x) = x^3`.




```python
# Crear una lista 'xs' que contiene valores de x desde -10 hasta 10
xs = [x for x in range(-10, 11)]

# Calcular los valores correspondientes de f(x) = x^3 para cada valor de x en 'xs'
fx = [x*x*x for x in xs]

# Crear un gráfico de línea que muestra la función f(x) = x^3 en 'xs'
# con un color azul hexadecimal (#0349fc)
plt.plot(xs, fx, color='#0349fc')

# Agregar una cuadrícula al gráfico con líneas grises punteadas
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Mostrar el gráfico resultante
plt.show()
```

**Ejercicio 2**

Usando el set de datos de películas dado hacer un gráfico de torta que muestre la proporción de películas vistas y no vistas.

```python
# Extraer la columna 'vista' de un DataFrame llamado 'peliculas' (debe haber sido previamente creado)
vistas = peliculas['vista']

# Contar la cantidad de 'sí' y 'no' en la columna 'vista'
conteo_vistas = vistas.count('si')
conteo_no_vistas = vistas.count('no')

# Almacenar los conteos en una lista llamada 'conteo'
conteo = [conteo_vistas, conteo_no_vistas]

# Etiquetas para los sectores del gráfico de torta ('Vistas' y 'No vistas')
labels = ['Vistas', 'No vistas']

# Colores para los sectores del gráfico de torta (verde y rojo)
colores = ['#109166', '#7d0c3d']

# Crear un gráfico de torta que muestra la proporción de 'Vistas' y 'No vistas'
# Utiliza 'autopct' para mostrar los porcentajes con dos decimales
plt.pie(conteo, labels=labels, colors=colores, autopct=lambda pct: '{:.2f}%'.format(pct))

# Mostrar el gráfico resultante
plt.show()
```

Dada el siguiente set de datos misterioso que contiene una serie de puntos con formato `(x, y)`




```python
# Importar la biblioteca random para generar números aleatorios
import random

# Generar una lista llamada 'puntos' que contendrá 100 pares de coordenadas (x, y)
# Las coordenadas x e y se generan de forma aleatoria en el rango [0, 1)
# Utilizamos una lista por comprensión para crear la lista
puntos = [(random.uniform(0, 1), random.uniform(0, 1)) for _ in range(100)]

# Acceder a los primeros 10 elementos de la lista 'puntos' e imprimirlos
# Esto muestra una muestra de los primeros 10 puntos generados
puntos[:10]
```

Hacer un gráfico de dispersión que muestre todos los puntos.

```python
# Crear dos listas vacías, 'xs' y 'ys', para almacenar las coordenadas x e y respectivamente
# xs = list()
# ys = list()

# Iterar a través de la lista 'puntos' para extraer las coordenadas x e y
# for x, y in puntos:
#     # Agregar la coordenada x a la lista 'xs'
#     xs.append(x)
#     # Agregar la coordenada y a la lista 'ys'
#     ys.append(y)

# Usando listas por comprensión
xs = [x for x, _ in puntos]
ys = [y for _, y in puntos]

# Crear un gráfico de dispersión (scatter plot) utilizando las listas 'xs' y 'ys'
# Establecer el color de los marcadores en un tono de rojo (#b33a15) y usar marcadores circulares ('o')
plt.scatter(xs, ys, color='#b33a15', marker='o')

# Agregar una cuadrícula al gráfico con líneas de color gris oscuro ('#18181c')
# Las líneas son punteadas ('--') y tienen un ancho de línea de 0.5
plt.grid(color='#18181c', linestyle='--', linewidth=0.5)

# Mostrar el gráfico de dispersión resultante
plt.show()
```
