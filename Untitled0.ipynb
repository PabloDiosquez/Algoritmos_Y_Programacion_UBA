```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('classic')
```

```python
xs = np.linspace(1, 15, 30)
ys = np.sin(xs) + 0.1*np.random.randn(len(xs))
```

```python
plt.plot(xs, ys)
plt.show()
```

```python
plt.plot(xs, ys, 'o--', color='purple', linewidth=0.7, ms=5)
plt.show()
```

```python
plt.figure(figsize=(8,3))
plt.plot(xs, ys, 'go-', linewidth=2, ms=5)
plt.show()
```

```python
plt.figure(figsize=(8,3))
plt.plot(xs, ys, 'o-', color='purple' ,linewidth=2, ms=5, label='Component 1')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.legend(loc='upper right', fontsize=8)
plt.show()
```

```python
x = np.linspace(1, 15, 100)
y = np.sin(x)

plt.figure(figsize=(8,3))
plt.plot(xs, ys, 'o', color='purple' ,linewidth=3, label='Component 1')
plt.plot(x , y, color='green', ms=3, label='Component 2')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.legend(loc='upper right', fontsize=8)
plt.show()
```

```python
plt.figure(figsize=(8,3))
plt.plot(xs, ys, 'o', color='purple' ,linewidth=3, label='Data')
plt.plot(x , y, color='green', ms=3, label='Theory')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.legend(loc='upper right', fontsize=8, ncols=2)
plt.ylim(top=2)
plt.grid(color='gray', linestyle='--', linewidth=.5)
plt.show()
```
